```python
import urllib
urllib.urlretrieve('https://ccrma.stanford.edu/workshops/mir2014/audio/C%20strum%20chord_mono.wav', filename='c_strum.wav')

from essentia.standard import MonoLoader
x = MonoLoader(filename='/Users/patrickryan/Desktop/kidtrunks1.wav')()
start = len(x)/2 + len(x)/4
end = len(x)/2 + len(x)/4 + len(x)/80
x= x[start + len(x)/9:end + len(x)/8]
from IPython.display import Audio
Audio(x, rate=44100)
```

```python
from librosa.feature import melspectrogram
from librosa.display import specshow
import essentia
import numpy
import warnings
warnings.filterwarnings("ignore") # librosa.cqt returns DeprecationWarnings, for some reason
fs = 44100.0
import librosa

%pylab inline
fmin = librosa.midi_to_hz(36)
fmax = librosa.midi_to_hz(96)
C = librosa.cqt(x, sr=fs, fmin=fmin)
logC = librosa.logamplitude(C)
figure(figsize=(14, 7))
librosa.display.specshow(logC, sr=fs, x_axis='time', y_axis='cqt_note', fmin=fmin, fmax=fmax)
savefig('back6.png')
```

```python
from librosa.feature import melspectrogram

from librosa.display import specshow
S = melspectrogram(x, sr=fs, n_fft=1024)
logS = librosa.logamplitude(S)
figure(figsize=(14, 7))
specshow(logS, sr=fs, x_axis='time', y_axis='mel')
savefig('back7.png')
```

```python
C = librosa.feature.chromagram(x, fs)
figure(figsize=(14, 7))
librosa.display.specshow(C, y_axis='chroma')
savefig('back8.png')
```
