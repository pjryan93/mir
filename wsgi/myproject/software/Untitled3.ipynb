```python
import essentia
from scipy import fft
from librosa.beat import *
import essentia.standard as ess
from numpy import *
import wave as wv

class Song(object):
    def __init__(self,audioPath):
        self.wave_file = wv.open(audioPath, 'r')
        self.numberOfChannels = self.wave_file.getnchannels()
        self.fs = 44100.0
        self.audio = ess.MonoLoader(filename=audioPath)()
        self.audio = self.audio[0:len(self.audio-(len(self.audio)/4 * 3))]
        self.avg_bpm, self.beat_start, self.confidence, self.tempo, self.beat_duration = ess.RhythmExtractor2013(method='multifeature')(self.audio)
        self.segments = self.setSegments()
    def getSegmentsFromOnsets(self):
        timeIndex = 0;
        segments = list()
        segment = list()
        startTime = 0
        increment = 4
        endTime = self.beat_start[increment - 1]
        print len(self.audio)
        for frameIndex in range(0,len(self.audio)-2):
            time = float(frameIndex)/float(self.fs)
            if  startTime < time and time < endTime:
                segment.append(self.audio[frameIndex])
            elif time > endTime:
                segments.append(segment)
                segment = list()
                timeIndex= timeIndex + increment
                startTime = self.beat_start[timeIndex]
                if timeIndex < len(self.beat_start)-increment:
                    endTime = self.beat_start[timeIndex+increment]
                else:
                    break
            else:
                continue
        segments.append(segment)
        return segments
    def setSegments(self):
        segments = self.getSegmentsFromOnsets()
        segments = array(segments)
        for i in segments:
            i = array(i)
        return segments
audioFilePath = "/Users/patrickryan/Desktop/untitled folder 2/data/turn3.wav"
tempSong = Song(audioFilePath)
```
